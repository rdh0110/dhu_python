```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd

driver = webdriver.Chrome()
driver.get("https://realty.daum.net/")

xpath = '//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[1]'
element = driver.find_element(By.XPATH, xpath)
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", element)

# 드롭다운에서 시/도 값 선택하기

city_list = '//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[2]'
driver.find_element(By.XPATH, city_list).click()

city = "대구광역시"

city_element=f'//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[2]/div[2]/div/div//*[contains(text(), "{city}")]'

WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, city_element)))

driver.find_element(By.XPATH, city_element).click()

# 시/군/구 값 선택하기

town= ""

if town:
    town_list = '//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[3]'
    driver.find_element(By.XPATH, town_list).click()

    town_element = f'//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[3]/div[2]/div/div//*[contains(text(), "{city}")]'

    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, city_element)))

    driver.find_element(By.XPATH, city_element).click()
time.sleep(1)

# 인기단지 데이터 수집하기

ranking = driver.find_elements(By.XPATH, "//*[contains(@class, 'sc-dcb40e1a-21')]")

#print(ranking[0].text)

ranking = driver.find_elements(By.XPATH, "//*[contains(@class, 'sc-dcb40e1a-21')]")[0:3]

# 개별 데이터 수집하기
#수집한 데이터를 데이터프레임으로 변환하기
rows=[]
#for data in ranking:
#    find_data = data.find_elements(By.CLASS_NAME, 'css-1563yu1')

#    row = {
#        '아파트명' : find_data[0].text,
#        '정보' : find_data[1].text,
#        '순위' : find_data[2].text
#}
#    rows.append(row)

#df= pd.DataFrame(rows)
#df

image_elements = driver.find_elements(By.CLASS_NAME, "sc-dcb40e1a-20")[0:3]

image_urls= []

for image_element in image_elements:
    style_attribute = image_element.get_attribute("style")
    image_url = style_attribute.split('url("')[1].split('")')[0]
    image_urls.append(image_url)

rows=[]
for data in ranking:
    find_data = data.find_elements(By.CLASS_NAME, 'css-1563yu1')

    row = {
        '아파트명' : find_data[0].text,
        '정보' : find_data[1].text,
        '순위' : find_data[2].text
}
    rows.append(row)

df= pd.DataFrame(rows)
df["이미지"] = image_urls

#df["이미지"][0] <- 확인해보기

df
```
